```python
import matplotlib.pyplot as plt
import numpy as np

years = ['2021', '2023']
native = [7, 4]
non_native = [9, 7]

x = np.arange(len(years))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 6))
rects1 = ax.bar(x - width/2, native, width, label='Native', color='#e9b87f')
rects2 = ax.bar(x + width/2, non_native, width, label='Non-native', color='#3896c4')

ax.set_ylabel('No. of Birds')
ax.set_title('Birds in Rizal Park by Type (Nov)')
ax.set_xticks(x)
ax.set_xticklabels(years)
ax.legend()

ax.set_yticks(np.arange(0, max(max(native), max(non_native)) + 1, 1))

plt.show()
```

```python
import numpy as np

# Data for native species
years = np.array([2021, 2023])  # X values
native_counts = np.array([7, 4])  # Y values for natives

# Calculate slope (b1) and intercept (b0) using the formula
b1_native = (np.sum((years - np.mean(years)) * (native_counts - np.mean(native_counts)))) / np.sum((years - np.mean(years))**2)
b0_native = np.mean(native_counts) - b1_native * np.mean(years)

# Predict for 2025
year_to_predict = 2025
predicted_native = b0_native + b1_native * year_to_predict

print(f"Native: Slope = {b1_native}, Intercept = {b0_native}, Prediction for 2025 = {predicted_native}")

# Data for non-native species
nonnative_counts = np.array([9, 7])  # Y values for non-natives

# Calculate slope (b1) and intercept (b0)
b1_nonnative = (np.sum((years - np.mean(years)) * (nonnative_counts - np.mean(nonnative_counts)))) / np.sum((years - np.mean(years))**2)
b0_nonnative = np.mean(nonnative_counts) - b1_nonnative * np.mean(years)

# Predict for 2025
predicted_nonnative = b0_nonnative + b1_nonnative * year_to_predict

print(f"Non-native: Slope = {b1_nonnative}, Intercept = {b0_nonnative}, Prediction for 2025 = {predicted_nonnative}")
```
